```python
import seaborn as sb
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu =0
sigma = 1
#dados de 0 a 100
dados = np.linspace(-100,100,1000)
```

```python


pdf = norm.pdf(dados,loc=mu,scale=sigma)
#print(pdf)
#distribuicao de função acumulada
cdf = norm.cdf(dados,loc=mu,scale=sigma)
```

```python
#normalizar dados
mu,sigma=norm.fit(dados)
dados_normal= mu,sigma
print(dados_normal)
```

```python
#dados aleatórios
n = 1000
amostra = norm.rvs(mu,sigma,n)
```

```python
sb.histplot(amostra,kde=True)
plt.show()
```

```python
plt.plot(dados, pdf, 'k', linewidth=2)
plt.title('Histograma e PDF')
```

```python
plt.plot(dados, cdf, 'k', linewidth=2)
plt.title('CDF')
```
