```python
import seaborn as sb
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu =0
sigma = 1
#dados aleat√≥rios
dados = np.linspace(-100,100,100)
print(dados)
```

```python
pdf = norm.pdf(dados,loc=mu,scale=sigma)
#print(pdf)
cdf = norm.cdf(dados,loc=mu,scale=sigma)
print(cdf)
```
