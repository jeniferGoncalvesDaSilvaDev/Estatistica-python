```python
import seaborn as sb
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu =0
sigma = 1
#dados de 0 a 100
dados = np.linspace(0,100,100)
print(dados)
```

```python

x=0
pdf = norm.pdf(x,loc=mu,scale=sigma)
#print(pdf)
#distribuicao de função acumulada
cdf = norm.cdf(0,loc=mu,scale=sigma)
print(pdf,cdf)
```

```python
#normalizar dados
mu,sigma=norm.fit(dados)
dados_normal= mu,sigma
print(dados_normal)
```

```python
#dados aleatórios
n = 100
amostra = norm.rvs(mu,sigma,n)
```

```python
sb.histplot(amostra,kde=True)
plt.show()
```

```python
plt.plot(x, pdf, 'k', linewidth=2)
plt.title('Histograma e PDF')
```
