```python
import seaborn as sb
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu =0
sigma = 1
#dados aleatórios
dados = np.linspace(0,100,100)
print(dados)
```

```python
pdf = norm.pdf(dados,loc=mu,scale=sigma)
#print(pdf)
#distribuicao de função acumulada
cdf = norm.cdf(dados,loc=mu,scale=sigma)
```

```python
#normalizar dados
mu,sigma=norm.fit(dados)
print(mu,sigma)
```

```python
sb.histplot(dados,kde=True)
plt.show()
```
